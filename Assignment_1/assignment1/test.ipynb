```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Step 1: Read the grayscale image
image = cv2.imread("D:/MBZUAI 2022-2027/MBZUAI 2024-2027/Semester 2/CV804/Exercise 1/CV804-2025-Spring-Semester-Starter-Code/assignment1/assets/depth_data/OhioPyle-256.jpg", cv2.IMREAD_GRAYSCALE)

# Step 2: Normalize the image (if necessary)
height_map = image / 255.0  # Normalize to range 0-1

# Step 3: Create a meshgrid for X, Y coordinates
# We create the X, Y grid based on the image size
x = np.arange(0, height_map.shape[1])
y = np.arange(0, height_map.shape[0])
x, y = np.meshgrid(x, y)

# Step 4: Plotting the 3D surface
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(x, y, height_map, cmap='viridis', edgecolor='none')

# Labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Height')
ax.set_title('Height Map')

plt.show()
```
